```python
linked_list = (5,1,(1,(9,None)))


def head(l):  #returns the head of list l
    return l[0]
def tail(l): #returns the tail of list l
    return l[1]
def cons(e,l): #return the list e::1
    return (e,l)
def length(l): #return the length of the list
    if l == None:
        return 0
    else:
        return length(tail(l)) + 1
    
#Below is a more efficient algorithm to compute length(l) by making the function tail recursive

def helper_length(n,l):
    #return n + length of l
    if l == None:
        return n
    else:
        return helper_length(n+1,tail(l))
def newlength(l):
    return helper_length(0,l)

# End efficient length function

def nth(l,n):     # Find the nth element in a linked list by recursively calling nth on the tail and subtracting the specified index n
    if n == 0:
        return head(l)
    else: 
        return nth(tail(l), n - 1)

def addlast(e,l):
    if l == None:
        return cons(e,None)
    else:
        return cons(head(l),addlast(e,tail(l)))

def reverse(l):
    if l == None:
        return None
    else:
        addlast(head(l), reverse(tail(l)))
        
def concat(l1,l2):
    if l1 == None:
        return l2
    else:
        return cons((head(l1)),concat(tail(l1),l2))

def concats(ls):
    if ls == None:
        return None
    else:
        return concat(head(ls),concats(tail(ls)))

def map_(f,l):
    if l == None:
        return None
    else:
        return cons(f(head(l)),map_(f,tail(l)))
    
def foldleft(f,start,l):
    if l == None:
        return start
    else:
        return foldleft(f,f(start,head(l)),tail(l))
    
def forall(p,l): #check if all elements of list l satisfy predicate p 
    return foldleft(lambda x,y: x and y,True,map_(p,l))

def exists(p,l):
    return foldleft(lambda x,y: x or y,False,map_(p,l))

def mem(e,l):
    return exists(lambda x: x==e,l)

def sort(l):
    return foldleft(insert,None,l)

def insert(l,e):
    if l == None:
        return cons(e,None)
    else:
        if e < head(l):
            return cons(e,l)
        else:
            return cons(head(l),insert(tail(l),e))
     
```
