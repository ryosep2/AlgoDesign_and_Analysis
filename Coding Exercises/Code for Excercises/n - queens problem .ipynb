```python
def check(queens):
    for i, (r1,c1) in enumerate(queens):
        for j, (r2,c2) in enumerate(queens):
            if i == j:
                continue
            if attacks((r1,c1),(r2,c2)):
                return False

    return True

def attacks(queen1, queen2):
    (r1,c1) = queen1
    (r2,c2) = queen2
    return r1 == r2 or c1 == c2 or abs(r1-r2) == abs(c1-c2)
```

```python
def findqueens(n):
    def helper(queens):
        """ Returns:
        a list of k queens placed on the
        last k rows of an nxn chessboard, if it exists
        None, otherwise"""
        if len(queens) == n:
            if check(queens):
                return queens
            else:
                return None
        
    
    return helper(n)
```

```python
import numpy as np

def decompose(n):
    a = []
    return decomposehelper(n,1,a)

def decomposehelper(n,digit,a):
    b = []
    if n <= 0:
        return None
    if digit > n:
        return a
    
    if n - digit == 0:
        b.append(digit)
        return a
    if np.mod(n,digit) == 0:
        if np.mod(n-digit-1,digit-1):
            b.append(digit)
```
