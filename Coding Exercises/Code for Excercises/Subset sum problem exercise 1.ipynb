```python
def subsetsum(s,t):
    if len(t) == 0 and s == 0:
        return t
    elif len(t) == 0 and s != 0:
        return None
    
    x = subsetsum(s-t[0],t[1:])
    if x == None:
        return subsetsum(s, t[1:])
    else:
        return [t[0]] + x
    
```

```python
def helper(n, start, temporary, end):
    if n == 0:
        return 
    helper(n-1,start,end, temporary)
    print('{} -> {}'.format(start,end))
    helper(n-1,temporary,end,start)
    
def hanoi(n):
    helper(n,1,2,3)
```
