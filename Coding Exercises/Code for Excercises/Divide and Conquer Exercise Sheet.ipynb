```python
# Q1. Brute force complexity is O(n^2) 
# when we don't care if i is less than k

# When we care if i < j, the complexity is n(n-1)/2


def buyandsell(tab):
    # base case: return the first and second indices of tab
    if len(tab) == 2:
        return (0,1)
    
    t1 = tab[:len(tab)//2] # size ~ n/2
    t2 = tab[len(tab)//2:] # size ~ n/2
    
    (b1,s1) = buyandsell(t1) 
    (b2,s2) = buyandsell(t2)
    (b2,s2) = (b2 + len(tab)//2, s2 + len(tab)//2) # O(1)
    
    b3 = t1.index(min(t1)) # O(n)
    s3 = t2.index(max(t2)) # O(n)
    s3 = s3 + len(tab)//2 #O(1)
    
    # therefore f(n) is O(n) due to it being the highest complexity 
    
    gain1 = t1[s1] - t1[b1]
    gain2 = tab[s2] - tab[b2]
    gain3 = tab[s3] - tab[b3]
    
    if gain1 == max(gain1,gain2,gain3):
        return (b1,s1)
    elif gain2 == max(gain1,gain2,gain3):
        return (b2,s2)
    else:
        return (b3,s3)
    
    # T(n) = O(n*logn) 
    
# def find_max_sum(tab):
```
